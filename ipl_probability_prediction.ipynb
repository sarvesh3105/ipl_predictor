```python
import pandas as pd
import numpy as np

match=pd.read_csv("matches.csv")
delivery=pd.read_csv("deliveries.csv")
```

```python
match.head()
```

```python
delivery.head()
```

```python
total_score_df=delivery.groupby(['match_id','inning'])['total_runs'].sum().reset_index()
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
match.rename(columns={'id':'match_id'},inplace=True)
```

```python
match.columns
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],on='match_id')
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
teams={'Sunrisers Hyderabad', 
       'Mumbai Indians','Royal Challengers Bangalore',
       'Kolkata Knight Riders','Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals','Delhi Capitals'}

match_df['team1']=match_df['team1'].replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
match_df['dl_applied'].unique()
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df['dl_applied'].unique()
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.head()
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.head()
```

```python
delivery_df.shape
```

```python
delivery_df['total_runs']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['total_runs']+1
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
delivery_df.head()
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else 1)
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype(int)
```

```python
delivery_df['wickets']=10-delivery_df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

```python
delivery_df['crr']=(delivery_df['total_runs']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
delivery_df['rr']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
def winning_team(x):
    if x['batting_team']==x['winner']:
        return 1
    else:
        return 0

delivery_df['result']=delivery_df.apply(winning_team,axis=1)
```

```python
delivery_df.columns
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','winner','total_runs_x','runs_left','balls_left','wickets','crr','rr','result']]
```

```python
final_df.head()
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.head()
```

```python
final_df=final_df[final_df['balls_left']!=0]
final_df
```

```python
X=final_df.drop(columns=['result'])
y=final_df['result']
X.drop(columns='winner',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
step1=ColumnTransformer(transformers=[('trf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,2,3])],
                        remainder='passthrough')
step2=LogisticRegression()

pipe=Pipeline([
    ('step1',step1),
    ('step2',step2)])

pipe.fit(X_train,y_train)

y_pred=pipe.predict_proba(X_test)

print("accuracy score",accuracy_score(y_test,y_test))
```
